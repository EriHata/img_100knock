```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Nearest Neighbor interpolation
def nn_interpolation(img, ax=1, ay=1):
    H, W, C = img.shape
    
    aH = int(ay*H)
    aW = int(ax*W)
    
    y = np.arange(aH).repeat(aW).reshape(aH, -1)  # 要素を繰り返す
    x = np.tile(np.arange(aW), (aH, 1))  # ndarrayを繰り返し並べる
    y = np.round(y/ay).astype(np.int)  # 四捨五入
    x = np.round(x/ax).astype(np.int) # 四捨五入
    
    # 上で添え字を作って
    # ここで指定しているのか？
    out = img[y,x]
    
    out = out.astype(np.uint8)
    
    return out

img = cv2.imread('../imori.jpg').astype(np.float)
plt.imshow(img.astype(np.uint8))
plt.show()
out = nn_interpolation(img, ax=1.5, ay=1.5)


plt.imshow(out)
cv2.imwrite('out25.jpg', out)
```
