## ヒストグラム正規化

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

# histgram normalization
def hist_normalization(img, a=0, b=255):
    # get max, min
    c = img.min()
    d = img.max()
    
    out = img.copy()
    
    # notmalization 
    out = (b-a)/(d-c)*(out-c) + a
    out[out<a] = a
    out[out>b] = b
    out = out.astype(np.uint8)
    
    return out

img = cv2.imread('../imori_dark.jpg').astype(np.float)
H, W, C = img.shape

#plt.imshow(img)

out = hist_normalization(img)

plt.hist(out.ravel(), bins=255, rwidth=0.8, range=(0, 255))
plt.savefig('out_hist.png')
plt.show()

# Save result
plt.imshow(out)
```
