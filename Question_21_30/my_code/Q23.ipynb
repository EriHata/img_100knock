## ヒストグラム平坦化

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline

def hist_equal(img,z_max=255):
    H, W, C = img.shape
    S = H*W*C*1.  # Sは画素数の総数
    
    out = img.copy()
    
    sum_h = 0.
    
    for i in range(1, 255):
        ind = np.where(img==i)
        sum_h += len(img[ind])
        z_prime = z_max/S*sum_h
        out[ind] = z_prime
        
    out = out.astype(np.uint8)
    return out


img = cv2.imread('../imori_dark.jpg')
out = hist_equal(img, z_max=255)

plt.hist(out.ravel(), bins=255, rwidth=0.8,range=(0,255))
plt.show()
plt.savefig('out_hist23.jpg')

plt.imshow(out)
```
