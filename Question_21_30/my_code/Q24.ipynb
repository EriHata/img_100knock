## ガンマ補正

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline

# gamma correction
def gamma_correction(img, c=1, g=2.2):
    out = img.copy()
    # [0,1]の間に正規化
    out /= 255.
    out = (1/c*out)**(1/g)
    out *= 255
    out = out.astype(np.uint8)
    
    return out

img = cv2.imread('../imori_gamma.jpg').astype(np.float)
# np.float型で、0-255の値を表示させようとしてもエラーになるのでintに修正
plt.imshow(img.astype(np.uint8))
plt.show()
out = gamma_correction(img, c=1, g=2.2)

plt.imshow(out)
cv2.imwrite('out24.jpg', out)
```
