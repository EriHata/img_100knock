## 平均プーリング

```python
import numpy as np
import cv2
%matplotlib inline
import matplotlib.pyplot as plt

def average_pooling(img, G=8):
    out = img.copy()
    
    # 画像の縦横Cを取ってくる
    H, W, C = img.shape
    
    Nh = int(H/G)
    Nw = int(W/G)
    
    for x in range(Nw):
        for y in range(Nh):
            for c in range(C):
                out[x*G:(x+1)*G, y*G:(y+1)*G, c] = np.mean(out[x*G:(x+1)*G, y*G:(y+1)*G, c]).astype(np.int)
                
    return out
    
    
img = cv2.imread('../imori.jpg')
out = average_pooling(img)
cv2.imwrite('out7.jpg', out)
plt.imshow(out)
```
