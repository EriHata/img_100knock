## 減色処理
ここでは画像の値を256^3から4^3、すなわちR,G,B in {32, 96, 160, 224}の各4値に減色せよ。

```python
import numpy as np
import cv2
%matplotlib inline
import matplotlib.pyplot as plt


def dicrease_color(img):
    img[(0<=img)&(img<64)]=32
    img[(64<=img)&(img<128)]=96
    img[(128<=img)&(img<192)]=160
    img[(192<=img)&(img<256)]=224
    
    return img

img = cv2.imread('../imori.jpg')
out = dicrease_color(img)

plt.imshow(out)
cv2.imwrite('out6-1.jpg', out)
```
