## アルファブレンド

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def alpha_blend(img1, img2, alpha=0.5):
    out = img1*alpha + img2 *(1-alpha)
    out = out.astype(np.uint8)
    return out

# Read image
img1 = cv2.imread("../imori.jpg").astype(np.float32)
img2 = cv2.imread('../thorino.jpg').astype(np.float32)

out = alpha_blend(img1, img2, alpha=0.6)

plt.imshow(out)

cv2.imwrite("out60.png", out)
```
